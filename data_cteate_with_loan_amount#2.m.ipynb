```python
df_l1=df_l[["CIFNO","LNAMOUNT"]]
```

```python
sum_by_cif = df_l1.groupby('CIFNO')['LNAMOUNT'].sum().reset_index()
```

```python
sum_by_cif.shape
```

```python
unique_cifno_count_lnamount = sum_by_cif['CIFNO'].nunique()

if unique_cifno_count_lnamount== len(sum_by_cif):
    print("All CIFNO values are unique.")
else:
    print("There are duplicate CIFNO values.")
```

```python
sum_by_cif.head()
```

```python
# Merge based on CIFNO and ACCOUNT_NO
merged_df = pd.merge(df, sum_by_cif, on=['CIFNO'], how='left')
```
